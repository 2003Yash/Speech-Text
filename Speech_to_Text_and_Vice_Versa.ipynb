Text to Speech

```python
# create environment by importing libraries
!pip install gtts # gtts = Google Text to Speech

from gtts import gTTS
```

```python
# define language and text
language = "en"
Text = "Hello, my name is Yaswanth and today i had my breakfast with batman"
```

```python
# create speech from given text
speech = gTTS(text = Text, lang = language, slow = False) #slow means voice will read out the text very slowly
                                                          # we can also change the voice and accent with more paramenters in same above function
```

```python
# save the speech as a file
speech.save("Text.mp3")
```
